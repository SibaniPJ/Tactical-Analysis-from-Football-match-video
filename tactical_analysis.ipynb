```python
import gradio as gr
from ultralytics import YOLO

model = YOLO("/content/runs/detect/train/weights/best.pt")

def predict(img):
    results = model(img)
    return results[0].plot()

demo = gr.Interface(fn=predict, inputs="image", outputs="image", title="Tactical Analysis Detector")
demo.launch(share=True)
```
